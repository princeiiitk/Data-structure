Given a list of integers nums and an integer target, write a function to return the indices of the two numbers such that they add up to the target.

```python
def Twosum(nums,target):
  dict1={}
  for i in range(len(nums)):
    rem = target - nums[i]
    nums[i]
    if rem in dict1:
      return (dict1[rem], i)
    dict1[nums[i]] = i

  return -1
if __name__=="__main__":
  print("enter size of list")
  n=int(input())
  nums=[]
  print("enter list value")
  for i in range(n):
    a=input()
    nums.append(int(a))
  print("enter target value")
  target=int(input())
  print(Twosum(nums,target))
```

2. Count Frequency of Elements in a List

Given a list of elements, write a Python function to count the frequency of each element and return the result as a dictionary where keys are elements and values are their counts.
Also, handle cases where the list contains mixed data types.

```python
def CountFreq(nums):
  dict1={}
  for i in nums:
    if i in dict1:
      dict1[i]=dict1[i]+1
    else:
      dict1[i]=1
  return dict1
if __name__=="__main__":
  print("enter size of list")
  n=int(input())
  nums=[]
  print("enter element of list")
  for i in range(n):
    x=input()
    nums.append(x)
  print("count frequency count",CountFreq(nums))
```
