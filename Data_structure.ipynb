2. Count Frequency of Elements in a List

Given a list of elements, write a Python function to count the frequency of each element and return the result as a dictionary where keys are elements and values are their counts.
Also, handle cases where the list contains mixed data types.

```python
def CountFreq(nums):
  dict1={}
  for i in nums:
    if i in dict1:
      dict1[i]=dict1[i]+1
    else:
      dict1[i]=1
  return dict1
if __name__=="__main__":
  print("enter size of list")
  n=int(input())
  nums=[]
  print("enter element of list")
  for i in range(n):
    x=input()
    nums.append(x)
  print("count frequency count",CountFreq(nums))
```

3. Find the First Non-Repeating Character

Given a string, write a function to find the first character that does not repeat.
Use a dictionary to track character frequencies and return the first unique character.
If all characters repeat, return a suitable message or value.

```python
def First_Non_Char(str1):
  dict1={}
  for i in str1:
    if i in dict1:
      dict1[i]=dict1[i]+1
    else:
      dict1[i]=1
  for key,value in dict1.items():
    if value==1:
      return key
  return -1
if __name__=="__main__":
  print("enter string")
  str1=input()
  print(First_Non_Char(str1))
```

4. Group Anagrams Using Dictionary

Given a list of strings, group the words that are anagrams of each other.
Return the result as a list of lists, where each inner list contains words that are anagrams.
Use a dictionary-based approach to efficiently group similar patterns.

```python
def AnagramCheck(str1):
  dict2={}

  for word in str1:
    key=''.join(sorted(word))
    if key not in dict2:
      dict2[key]=[]
    else:
      dict2[key]=word
  return list(dict2.values())

if __name__=="__main__":
  print("enter size of list")
  n=int(input())
  print("enter string")
  str1=[]

  for i in range(n):
    x=input()
    str1.append(x)
  print(AnagramCheck(str1))
```

Combination Sum
Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the frequency of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.



Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []


```python
def combinationSum(idx,nums,target,ans,temp):
  if target==0:
      ans.append(temp.copy())
      return
  if len(nums)==idx:
    return
  if target>=nums[idx]:
    temp.append(nums[idx])
    combinationSum(idx,nums,target-nums[idx],ans,temp)
    temp.pop()
  combinationSum(idx+1,nums,target,ans,temp)
  return

if __name__=="__main__":

  nums=[2,2,3,8,5]
  ans=[]
  temp=[]
  target=8
  combinationSum(0,nums,target,ans,temp)
  print(ans)

```
